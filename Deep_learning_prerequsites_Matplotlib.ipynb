# Matplotlib
Line Chart  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.linspace(0,20,1000)
Y= np.sin(X)+.2*X
```

```python
plt.plot(X,Y)
plt.show()
```

```python
plt.plot(X,Y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('sin(x)+.2x')
plt.show()
```

```python
x = np.random.randn(200,2)
x[:50]+=3
Y=np.zeros(200)
print(Y)
Y[:50]=1
Y[50:70]=2
print(Y)
plt.scatter(x[:,0],x[:,1],c=Y,cmap='brg')
plt.show()
print(x)
```
