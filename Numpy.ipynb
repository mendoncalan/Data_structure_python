The Dot Product

```python
a = np.array([1,2])
```

```python
b= np.array([3,4])
```

```python
dot =0
for e , f in zip(a,b): # zip() function is a python built in function for combining Multiple iterable object like List, tuples or string into a single iterable Tuples. https://www.codecademy.com/article/python-zip-function
  dot+=e*f
```

```python
dot
```

```python
dot=0
for i in range(len(a)):
  dot+= a[i]*b[i]
```
